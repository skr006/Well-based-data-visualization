```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(
    "../Backend/data/wells.csv",
    parse_dates=["DATEPRD"]
)

well_name = "15/9-F-12"
df_well = df[df["NPD_WELL_BORE_NAME"] == well_name]

df_well = df_well.dropna(subset=["AVG_DOWNHOLE_PRESSURE"])

df_well = df_well.sort_values("DATEPRD")

plt.figure(figsize=(10, 5))
plt.plot(
    df_well["DATEPRD"],
    df_well["AVG_DOWNHOLE_PRESSURE"]
)

plt.xlabel("Date")
plt.ylabel("Average Downhole Pressure")
plt.title(f"AVG_DOWNHOLE_PRESSURE vs Date ({well_name})")
plt.grid(True)
plt.tight_layout()
plt.show()
```
